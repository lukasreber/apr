```python
def random_dist(lst,value):
    i = 0
    if len(lst) != 0:
        while i < len(lst):
            if lst[i][0] == value:
                lst[i] = (value,lst[i][1]+1)
                return lst
            i += 1
    lst.append((value,1))
    return lst
        
```

```python
# record complexity
def save_runtime(n):
    i = 0
    l = []
    while i < n*10:
        l = random_dist(l, random.randint(0,n))
        i += 1
    return l
```

```python
save_runtime(10)
```

```python
times = []
y = 1000
for n in range(0,y):
    TEST_CODE = f'save_runtime({n})'
    times.append(timeit.timeit(globals=globals(),stmt=TEST_CODE,number=1))
```

```python
x = [*range(0,y)]
plt.plot(x,times)
plt.xlabel('Input size')
plt.ylabel('Time')
```
